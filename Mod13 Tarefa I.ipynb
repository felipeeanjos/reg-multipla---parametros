```python
reg1 = smf.ols('np.log(renda) ~  sexo + posse_de_veiculo + posse_de_imovel + C(qtd_filhos) + tipo_renda + educacao + estado_civil + tipo_residencia + idade + tempo_emprego + C(qt_pessoas_residencia,Treatment(2))  ',df).fit()
```

```python
reg1.summary()
```

```python
reg1_summary = {
    'r²':'0.358',
    'r²a':'0.357',
    'aic':'2.719'
}
```

```python
reg2 = smf.ols('np.log(renda) ~ sexo + posse_de_veiculo + posse_de_imovel + C(qtd_filhos) + estado_civil + idade + tempo_emprego + C(qt_pessoas_residencia,Treatment(2))  ',df).fit()
```

```python
reg2.summary()
```

```python
reg2_summary = {
    'r²':'0.348',
    'r²a':'0.347',
    'aic':'2.736'
}
```

```python
reg3 = smf.ols('np.log(renda) ~ sexo + posse_de_veiculo + posse_de_imovel + idade + tempo_emprego ',df).fit()
```

```python
reg3.summary()
```

```python
reg3_summary = {
    'r²':'0.346',
    'r²a':'0.346',
    'aic':'2.736'
}
```

```python
print('modelo1:')
print('r²: '+reg1_summary['r²'])
print('r² ajustado: '+reg1_summary['r²a'])
print('AIC: '+reg1_summary['aic'])
print('\n')

print('modelo2:')
print('r²: '+reg2_summary['r²'])
print('r² ajustado: '+reg2_summary['r²a'])
print('AIC: '+reg2_summary['aic'])
print('\n')


print('modelo3:')
print('r²: '+reg3_summary['r²'])
print('r² ajustado: '+reg3_summary['r²a'])
print('AIC: '+reg3_summary['aic'])
print('\n')
```

Se compararmos os modelos podemos ver que os parâmetros caíram, entretando, diminuímos consideravelmente o número de variáveis para uma caída bem baixa nos parâmetros, analisando essas informações e levando em conta que removemos todas as variáveis menos significantes, eu prefiro optar por manter o modelo 3, caindo bem pouco nesses parâmetros mas levando em conta a análise de um modelo bem mais 'leve'
